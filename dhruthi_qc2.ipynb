```python
from qiskit import *
import matplotlib as mp
```

```python
input = QuantumRegister(2)
qr = QuantumRegister(2,2)
carry = ClassicalRegister(1)
sum = ClassicalRegister(1)
circuit = QuantumCircuit(input,qr,carry,sum)
```

```python
def add(circuit,input,qr,cr1,cr2):

  circuit.cx(input[0],qr[0])
  circuit.cx(input[1],qr[0])
  circuit.ccx(input[0],input[1],qr[1])
  circuit.measure(qr[0],sum)
  circuit.measure(qr[1],carry)
```

```python
add(circuit,input,qr,carry,sum)
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend=simulator).result()
plot_histogram(result.get_counts(circuit))
```

```python
circuit.reset(qr)
circuit.x(input[1])
add(circuit,input,qr,carry,sum)
circuit.draw(output='mpl')
```

```python
result = execute(circuit, backend=simulator).result()
plot_histogram(result.get_counts(circuit))
```

```python
circuit.reset(qr)
circuit.reset(input)
circuit.x(input[0])

add(circuit,input,qr,carry,sum)
circuit.draw(output='mpl')
```
