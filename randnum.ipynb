```python
from qiskit import *
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=10
```

```python
val=ClassicalRegister(n)
qub=QuantumRegister(n)
qc=QuantumCircuit(qub,val)
for i in range(n):
    qc.h(i)
for i in range(n):
    qc.measure(i,i)
qc.draw()
```

```python
simulator = Aer.get_backend('qasm_simulator')
job = execute(qc, simulator, shots=1000)
result = job.result()
counts = result.get_counts(qc)
# plot_histogram(result.get_counts(qc))
# print(counts)
measurement_values = [int(key, 2) for key in counts.keys()]


# Create a NumPy array with the measurement values
measurements_array = np.array(measurement_values)

print(measurements_array)
rand_num=0
rand_num=measurements_array[0]
print(rand_num)
```

```python
binary_dict = {}
for num in measurements_array:
    if num in binary_dict:
        binary_dict[num] += 1
    else:
        binary_dict[num] = 1

# Create histogram
plt.bar(binary_dict.keys(), binary_dict.values())
plt.xlabel('Decimal Number')
plt.ylabel('Count')
plt.title('Histogram of Decimal Numbers')
plt.show()
```
